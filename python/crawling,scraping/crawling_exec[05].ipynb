# 왜 크롤링이 필요한지 / 크롤링 준비 / 이미지 스크래핑 / 자연어 처리

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
from urllib.error import HTTPError
from urllib.error import URLError
import pandas as pd
from urllib.parse import quote_plus
```

```python
base_url = 'https://search.naver.com/search.naver?sm=t&where=image&query=' # base url 이다. 여기서 뒤에 어떤 검색어를 넣냐에 따라 달라진다.
keyword = input('검색어 입력 :') # 여기서 네이버에서 이미지 검색하던것 처럼 명령어를 입력해준다.
imgCnt = int(input('스크랩 할 이미지 개수 : ')) # 스크랩 할 이미지 개수를 지정해준다.
try:
    html = urlopen(base_url+quote_plus(keyword)) # quote_plus 를 통해 URL Encoding을 한다.
except HTTPError as he:
    print('http error')
except URLError as us:
    print('url error')
else:
    soup = BeautifulSoup(html.read(),'html.parser')
```

```python
img  = soup.find_all('img','_img')
```

```python
img
```

```python
cnt = 1 
for i in img:
    imgUrl = i['data-source']
#     print(imgUrl)
    with urlopen(imgUrl) as file: # 해당 이미지 파일을 저장하기 위한 로직이다.
        with open('../image/'+str(cnt)+'.jpg','wb')as imgFile: # 해당 파일을 해당 경로에 이름을 만들어 저장한다.wb는 write binary 라는 뜻이다.
            img = file.read() # 이미지 파일을 읽고
            imgFile.write(img) # 읽은 파일을 write로 저장한다.
    cnt+=1 # 각 이름을 바꾸기 위해 cnt+1 을 한다.
    if cnt>imgCnt: # 우리가 처음 지정했던 50개가 넘어가면 멈추게 break로 멈추게 한다.
        break
print('Image download success')
```

# 크롤링을 통한 페이지 처리 방법 익히기

```python
from time import sleep
from random import randint
import pandas as pd
from IPython.core.display import clear_output
from time import time
from bs4 import BeautifulSoup
from requests import get
```

```python
# loop 사이에 잘 진행되고 있는지를 보기 위한 indicator sentence
start_time=time()

requests=0
for _ in range(5):
    requests += 1
    sleep(randint(1,3)) # 간격을 둔다
    current_time = time() # 현재시간
    elapsed_time = current_time = start_time # 이 3개의 값을 같게하고 
    print("request:{},frequency :{}".format(requests,requests/elapsed_time)) # 시간을 체크한다.
    clear_output(wait=True)
```

```python
base_url = 'https://search.shopping.naver.com/search/all?frm=NVSHATC&origQuery=%ED%8C%8C%EC%9D%B4%EC%8D%AC&pagingIndex='+page+'&pagingSize=40&productSet=total&query=%ED%8C%8C%EC%9D%B4%EC%8D%AC&sort=rel&timestamp=&viewType=list' # base url 이다. 여기서 뒤에 어떤 검색어를 넣냐에 따라 달라진다.
keyword = input('검색어 입력 :') # 여기서 네이버에서 검색하던것 처럼 명령어를 입력해준다.

try:
    html = urlopen(base_url+quote_plus(keyword)) # quote_plus 를 통해 URL Encoding을 한다.
except HTTPError as he:
    print('http error')
except URLError as us:
    print('url error')
else:
    soup = BeautifulSoup(html.read(),'html.parser')
```

```python
pages = [ str(i) for i in range(1,11)]
pages
```

```python
# loop 사이에 잘 진행되고 있는지를 보기 위한 indicator sentence
start_time=time()
requests=0
prices_list=[]
title_list=[]
link_list=[]
for page in pages:
    requests += 1 # get이 open과 같은 역할이다.
    html = urlopen('https://search.shopping.naver.com/search/all?frm=NVSHATC&origQuery=%ED%8C%8C%EC%9D%B4%EC%8D%AC&pagingIndex='+page+'&pagingSize=40&productSet=total&query=%ED%8C%8C%EC%9D%B4%EC%8D%AC&sort=rel&timestamp=&viewType=list') # base url 이다. 여기서 뒤에 어떤 검색어를 넣냐에 따라 달라진다.
    sleep(randint(1,3)) # 간격을 둔다
    current_time = time() # 현재시간
    elapsed_time = current_time = start_time # 이 3개의 값을 같게하고 
    print("request:{},frequency :{}".format(requests,requests/elapsed_time)) # 시간을 체크한다.
    soup=BeautifulSoup(html.read(),'html.parser')
    # title, link , price
    # 가격 가져오기
    prices = soup.select('.price_num__2WUXn')
    prices_text = [price.text for price in prices]
    prices_list.extend(prices_text)
    # 타이틀, 링크 가져오기
    titles = soup.select('.basicList_link__1MaTN')
    titles_text = [title.text for title in titles]
    title_list.extend(titles_text)
    
    link_text = [title['href'] for title in titles]
    link_list.extend(link_text)
```

```python
print(len(prices_list))
print(len(title_list))
print(len(link_list))
```

```python
python_page_df = pd.DataFrame({'title':title_list,'price':prices_list,'link':link_list})
python_page_df
```

```python
python_page_df.to_csv('.python_page_df.csv',mode='w',encoding='utf-8',index=False)
print('success')
```
