# json 형식의 파일 크롤링

```python
import json, re
from urllib.request import urlopen
from html import unescape
```

```python
# 웹 페이지 읽어오기
request = urlopen('http://www.hanbit.co.kr/store/books/full_book_list.html')
encoding = request.info().get_content_charset('utf-8') # 인코딩 부분 셋팅하는 부분
html = request.read().decode(encoding)
html
```

- json.dumps() : 데이터를 json형태로 인코딩(문자열)
- json.dump() : 데이터를  json 형태로 인코딩하여 파일에 출력(파일)
- ensure_ascii = False:
- [{key : value}, {key : value}] indent = size

- 정규표현식
- .모든문자
- * 0번이상 반복
- ? 있어도 되고 없어도 되는
- <a hred="(.*>)">
- <td class = 'left'><a.*></td> # a태그로 시작해서 td 태그로 끝나는 것을 가져온다

###  json 파일생성

```python
with open('booklist_json.json', mode='w', encoding='utf-8') as file:
    data = []
    for partial_html in re.findall(r'<td class="left"><a.*?</td>', html): #앞에 r이라는 뜻은 로우데이터라는 뜻이다. 안그러면 < 이런게 바뀐다.
#         print(partial_html) # 또한 정규표현식에서 쓰는 re는 find_all이 아닌 findall 이다.
        search = re.search(r'<a href="(.*?)">',partial_html).group(1) # 그룹에 첫번째 요소에 대해서 가져온다.
        url = 'http://www.hanbit.co.kr' + search
#         print(url)
        title = re.sub(r'<.*?>','', partial_html)
        print(title)
        data.append({'bookName':title, 'link':url})
        print(json.dumps(data,ensure_ascii=False,indent=2))
    json.dump(data,file,ensure_ascii=False,indent=2)
```

- 전체 제조사 엘리먼트를 찾아서 마우스 클릭하고 그 결과물을 반환 받는다.

- 다음 페이지 버튼이  비활성화 되어 있으면 크롤링을 중단한다.

```python
bikeElementBtn = driver.find_element_by_css_selector('.detail_more')
# bikeElementBtn = driver.find_element_by_css_selector('#container > div.con')

bikeElementBtn.click()
```

```python
time.sleep(3)
```

```python
bikeElements = driver.find_elements_by_css_selector('#_vendor_select_layer > div > div.maker_group > div.emblem_area > ul > li')
# bikeElements = driver.find_elements_by_css_selector('#_vendor_select_layer>div>div.maker_group>div.emblem_area>ul>li')
bikeElements
```

```python
for bike_li in bikeElements:
    companyName = bike_li.find_element_by_tag_name('span').text
    print(companyName.strip())
    if companyName == '':
        break
    # 이미지
    img = bike_li.find_element_by_tag_name('img').get_attribute('src')
    print(img)
    link = bike_li.find_element_by_tag_name('a').get_attribute('href')
    print(link)
time.sleep(3)
    
```

```python
nextBtn = driver.find_element_by_css_selector('#_vendor_select_layer > div > div.maker_group > div.rolling_btn > button.next')
isFlag = nextBtn.is_enabled()
print(isFlag)
```

```python
if isFlag :
    print('next page exist~~~')
    nextBtn.click()
    bikeElements = driver.find_elements_by_css_selector('#_vendor_select_layer > div > div.maker_group > div.emblem_area > ul > li')
    print(bikeElements)
    for bike_li in bikeElements:
        companyName = bike_li.find_element_by_tag_name('span').text
        print(companyName.strip())
        if companyName == '':
            break
        # 이미지
        img = bike_li.find_element_by_tag_name('img').get_attribute('src')
        print(img)
        link = bike_li.find_element_by_tag_name('a').get_attribute('href')
        print(link)
```
